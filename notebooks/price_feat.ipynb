```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Датасет https://www.kaggle.com/datasets/olistbr/brazilian-ecommerce
path_to_datasets = "/content/drive/MyDrive"

orders = pd.read_csv(f"{path_to_datasets}/olist_order_items_dataset.csv")
purchases = pd.read_csv(f"{path_to_datasets}/olist_orders_dataset.csv", parse_dates=['order_purchase_timestamp'])
products = pd.read_csv(f"{path_to_datasets}/olist_products_dataset.csv")
```

```python
# Джоиним заказы, даты оплаты заказов и категории продуктов
dataset = orders[['order_id', 'product_id', 'price']].copy()\
    .merge(purchases[['order_id', 'order_purchase_timestamp']].copy(), how='left', on='order_id')\
    .merge(products[['product_id', 'product_category_name']].copy(), how='left', on='product_id')
```

```python
dataset.head()
```

```python
# Приводим дату оплаты заказа к дате начала недели
dataset['week'] = dataset['order_purchase_timestamp'].dt.to_period('W').dt.start_time
```

```python
# Товары, по которым у нас наибольшее колво исторической инфы и колва заказов
popular_positions = dataset.groupby(['product_id']).count().reset_index(drop=False).sort_values('week', ascending=False)
```

```python
# Выбираем товары, для которых есть записи, равные по количеству какому-то временному периоду
years = 1
weeks = 52 * years
popular_positions = popular_positions.groupby(['product_id']).count().reset_index(drop=False)\
    .sort_values('week', ascending=False)[popular_positions['week'] >= weeks]['product_id']
```

```python
# Итерируем товары из тех, которые мы отобрали
for prod in popular_positions:
    # Создаю датафрейм с данными по выбранному товару
    df = dataset[dataset['product_id'] == prod] \
        .groupby(['week']).agg(         # Группируем данные за неделю
        sum_price=('price', 'sum'),     # Суммарная цена
        max_price=('price', 'max'),     # Максимальная цена
        min_price=('price', 'min'),     # Минимальная цена
        mean_price=('price', 'mean'),   # Средняя цена
        quantity=('price', 'count')     # Суммарные продажи
    ).reset_index(drop=False)

    # За неимением лучшего полагаю себестоимость равной минимальной цене
    df['cost'] = df['min_price'].min()

    # Считаю показатели GMV, margin и markup
    df['gmv'] = df['mean_price'] * df['quantity']
    df['margin'] = (df['mean_price'] - df['cost']) / df['mean_price']
    df['markup'] = (df['mean_price'] - df['cost']) / df['cost']

    # Считаю эластичность за все время продаж, так как цена меняется очень редко
    sales_delta = df['quantity'].max() - df['quantity'].min()
    sales_avg = df['quantity'].mean()
    price_delta = df['max_price'].max() - df['cost'].min()
    price_avg = df['mean_price'].mean()

    # Так как для некоторых товаров цена остается постоянной, присваиваем их эластичности NA
    if price_delta != 0:
        df['elasticity'] = (sales_delta / sales_avg) / (price_delta / price_avg)
    else:
        df['elasticity'] = pd.NA
```
