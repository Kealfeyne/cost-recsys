```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import sys
import os
```

```python
project_root = os.path.abspath('..')
if project_root not in sys.path:
    sys.path.insert(0, project_root)
path_to_datasets = os.path.join(project_root, 'data', 'external', 'archive')
path_to_datasets
```

```python
orders = pd.read_csv(f"{path_to_datasets}/olist_order_items_dataset.csv")
purchases = pd.read_csv(f"{path_to_datasets}/olist_orders_dataset.csv", parse_dates=['order_purchase_timestamp'])
products = pd.read_csv(f"{path_to_datasets}/olist_products_dataset.csv")
orders.head()
```

```python
dataset = orders[['order_id', 'product_id', 'price']].copy()\
    .merge(purchases[['order_id', 'order_purchase_timestamp']].copy(), how='left', on='order_id')\
    .merge(products[['product_id', 'product_category_name']].copy(), how='left', on='product_id')
dataset.head()
```

```python
dataset['week'] = dataset['order_purchase_timestamp'].dt.to_period('W').dt.start_time
dataset.head()
```

```python
dataset.drop("order_purchase_timestamp", axis=1, inplace=True)
dataset
```

```python
dataset.drop("order_id", axis=1, inplace=True)
dataset
```

Category with the most purchases

```python
# Count the occurrences of each category
category_counts = dataset["product_category_name"].value_counts()

# Find the category with the most purchases
most_purchased_category = category_counts.idxmax()
dataset = dataset[dataset["product_category_name"] == most_purchased_category]
dataset
```

5 products with the most purchases in a filtered dataset

```python
# Count the occurrences of each product_id
product_counts = dataset["product_id"].value_counts()

# Get the top 5 product_ids with the most purchases
top_5_products = product_counts.head(5).index

# Filter the dataset to include only these top 5 products
top_5_dataset = dataset[dataset["product_id"].isin(top_5_products)]

# Display the result
top_5_dataset = top_5_dataset.reset_index(drop=True)

top_5_dataset
```

```python
top_5_dataset.drop("product_category_name", axis=1, inplace=True)
top_5_dataset
```

```python
top_5_dataset = (
    top_5_dataset
    .groupby(['week', 'product_id'])
    .agg(
        purchase_count=('price', 'size'),  # Count the number of rows (purchases)
        average_price=('price', 'mean')    # Calculate the average price
    )
    .reset_index()
)

top_5_dataset
```

Product_id encoding

```python
# Perform one-hot encoding for the 'product_id' column
one_hot_encoded = pd.get_dummies(top_5_dataset["product_id"], prefix="product")

# Combine the one-hot encoded data with the original dataset
top_5_dataset = pd.concat([top_5_dataset.reset_index(drop=True), one_hot_encoded], axis=1)

# Display the resulting dataset with one-hot encoding
top_5_dataset
```

```python
top_5_dataset.drop("product_id", axis=1, inplace=True)
top_5_dataset
```

```python
# Convert 'week' column to a datetime object if it's not already
top_5_dataset['week'] = pd.to_datetime(top_5_dataset['week'])

# Extract the day of the year (1-365) from the 'week' column
top_5_dataset['day_of_year'] = top_5_dataset['week'].dt.dayofyear

# Calculate angles for each day in a counterclockwise direction
top_5_dataset['angle'] = 2 * np.pi * (top_5_dataset['day_of_year'] / 365)

# Calculate x and y coordinates on the unit circle
top_5_dataset['x_coordinate'] = np.sin(top_5_dataset['angle'])
top_5_dataset['y_coordinate'] = np.cos(top_5_dataset['angle'])

# Drop intermediate 'angle' column if not needed
top_5_dataset_reset = top_5_dataset.drop("angle", axis=1, inplace=True)
top_5_dataset_reset = top_5_dataset.drop("day_of_year", axis=1, inplace=True)
top_5_dataset
```

```python
top_5_dataset['week'] = pd.factorize(top_5_dataset['week'])[0]

# Display the updated dataset
top_5_dataset
```
